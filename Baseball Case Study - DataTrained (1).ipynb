### VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

```python
def calculate_vif(dataset):
    X = add_constant(dataset)
    vif = pd.DataFrame()
    vif['Features'] = X.columns
    vif['VIF_Values'] = [variance_inflation_factor(X.values,i)for i in range(X.shape[1])]
    return(vif.sort_values(by='VIF_Values',ascending=False))
```

```python
calculate_vif(x)
```

Lets set the VIF threshold as 10.

We see that RA,ER and ERA have avery high correlation with each other and vif_ values are too high hence we could drop RA and ER

```python
x = x.drop(['ER','ERA'],axis=1)
```

```python
calculate_vif(x)
```

We see that all the values are within the threshold.

### Scaling the Model

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
X
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
from sklearn.linear_model import LinearRegression
```

### Base Model

```python
lr = LinearRegression()

for i in range(0,100):
    
    x_train,x_test,y_train,y_test = train_test_split(X,dfy,test_size=0.25,random_state=i)
    lr.fit(x_train,y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    print('At random_state',i,'the training accuracy is',r2_score(y_train,pred_train))
    print('At random_state',i,'the testing accuracy is',r2_score(y_test,pred_test))
    print()
```

Best Random State = 7

```python
x_train,x_test,y_train,y_test = train_test_split(X,dfy,test_size=0.25,random_state=7)
```

```python
lr.fit(x_train,y_train)
pred_test = lr.predict(x_test)
R2_Score = r2_score(pred_test,y_test)
MAE = mean_absolute_error(pred_test,y_test)
RMSE = np.sqrt(mean_squared_error(pred_test,y_test))

print('Model Report for :', lr)
print('r2_score is :' ,R2_Score)
print('MAE is :', MAE)
print('RMSE is :', RMSE)
```

```python
def Model_Reg(model):
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    print('Model Report')
    r2_Score = r2_score(y_test,pred)*100
    print('r2_score',r2_Score)
    print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
    print('Mean Absolute Error',mean_absolute_error(y_test,pred_test))
    coef = pd.Series(model.coef_,x.columns).sort_values(ascending=False)
    coef.plot(kind= 'bar', title = 'Model Coefficients')
```

```python
def Model_Tree(model):
    model.fit(x_train,y_train)
    pred = model.predict(x_test)
    print('Model Report')
    r2_Score = r2_score(y_test,pred)*100
    print('r2_score',r2_Score)
    print('RMSE',np.sqrt(mean_squared_error(y_test,pred)))
    print('Mean Absolute Error',mean_absolute_error(y_test,pred_test))
    coef = pd.Series(model.feature_importances_,x.columns).sort_values(ascending=False)
    coef.plot(kind= 'bar', title = 'Model Coefficients')
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso

lr = LinearRegression()
Model_Reg(lr)
```

```python
ridge = Ridge()
Model_Reg(ridge)
```

```python
lasso = Lasso()
Model_Reg(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dt = DecisionTreeRegressor()
Model_Tree(dt)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfr = RandomForestRegressor()
Model_Tree(rfr)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
et = ExtraTreesRegressor()
Model_Tree(et)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
gb = GradientBoostingRegressor()
Model_Tree(gb)
```

```python
from sklearn.linear_model import ElasticNet
en = ElasticNet()
Model_Reg(en)
```

We see that Linear Regression and Ridge are the best Models
